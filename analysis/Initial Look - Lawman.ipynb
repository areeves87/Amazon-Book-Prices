```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../data/cleaned_top_100_with_title_length.csv', encoding  = "ISO-8859-1")

df.head()
```

### Examine data

```python
#For these charts, select no audio books and no trial books
positive_cost_df = df[df['Price']>=0]
```

```python
categories_in_order_by_median_price = positive_cost_df.groupby('Category')['Price'].median().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(12,13))
plt.xticks(rotation=90)
sns.boxplot(x='Price',y='Category',data=positive_cost_df,orient='h', order = categories_in_order_by_median_price)
plt.title('CDs are have the highest median prices and romance books have the lowest median price');
```

### Examine Title lengths

```python
categories_in_order_by_median_length = positive_cost_df.groupby('Category')['Title Length'].median().sort_values(ascending=False).index
```

```python
plt.figure(figsize=(12,13))
plt.xticks(rotation=90)
sns.boxplot(x='Title Length', y='Category',data=positive_cost_df,orient='h',order=categories_in_order_by_median_length);
plt.title('Literature Fiction books have the shortest titles, Cookbooks have the longest');
```
